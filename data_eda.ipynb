```python
df_factors = df[[  'pneumonia', 'diabetes', 'copd',
       'asthma', 'inmsupr', 'hypertension', 'other_disease', 'cardiovascular',
       'obesity', 'renal_chronic', 'tobacco']]
```

```python
for i in df_factors.columns:
    fac_num = df_factors[i].value_counts()
    print("graph for %s: total = %d" % (i,len(fac_num)))
    chart = sns.barplot(x=fac_num.index, y=fac_num)
    chart.set_xticklabels(chart.get_xticklabels())
    plt.show()
```

```python
pd.set_option('display.max_rows', None)
pd.pivot_table(df,index = [  'pneumonia', 'diabetes', 'copd',
       'asthma', 'inmsupr', 'hypertension', 'other_disease', 'cardiovascular',
       'obesity', 'renal_chronic', 'tobacco'],values=['covid_res'])
```

```python
fig, ax = plt.subplots()

ax.hist(df[df["deceased"]==1]["age"], bins=15, alpha=0.5, color="blue", label="deceased")
ax.hist(df[df["deceased"]==0]["age"], bins=15, alpha=0.5, color="green", label="not deceased")

ax.set_xlabel("Age")
ax.set_ylabel("Count of patients")

fig.suptitle("Age vs. Survival for Covid Patients")

ax.legend();
```

```python
fig, ax = plt.subplots()

#ax.hist(df[df["covid_res"]==1]["age"], bins=15, alpha=0.5, color="blue", label="Positive")
ax.hist(df[df["covid_res"]==0]["age"], bins=15, alpha=0.5, color="red", label="Negative")

ax.set_xlabel("Age")
ax.set_ylabel("Count of patients")

fig.suptitle("Age vs. Negative Covid Results")

ax.legend();
```

```python
fig, ax = plt.subplots()

sns.kdeplot(df[df["deceased"]==1]["age"], shade=True, color="blue", label="Deceased", ax=ax)
sns.kdeplot(df[df["deceased"]==0]["age"], shade=True, color="green", label="Not Deceased", ax=ax)

ax.set_xlabel("Age")
ax.set_ylabel("Density")

fig.suptitle("Age vs. Deceased for Covid Patients");
```
